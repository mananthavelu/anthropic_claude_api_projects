```python
from dotenv import load_dotenv
load_dotenv()

from anthropic import Anthropic
client = Anthropic()
```

```python
# Initialize the model
messages = []
model = "claude-3-5-haiku-20241022"
```

```python
def add_user_message(messages, text):
    user_message = {"role": "user",
                    "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant",
                         "content": text}
    messages.append(assistant_message)

def chat(messages, system_prompt = None):
    model_params = {
        "model": model,
        "max_tokens": 1000,
        "temperature" : 0.0,
        "messages": messages,
        "stream" : True
    }
    if system_prompt:
        model_params["system_prompt"] = system_prompt
    response = client.messages.create(**model_params)
    for event in response:
        print(event)
```

```python
add_user_message(messages, "What is quantum computing")
chat(messages)
```

```python
model_params = {
        "model": model,
        "max_tokens": 1000,
        "temperature" : 0.0,
        "messages": messages
    }
```
